```python
#import the library used to query a website
import urllib.request #if you are using python3+ version, import urllib.request

#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup

urls_to_be_scrapped = ['https://en.wikipedia.org/wiki/Deep_learning']
```

```python
import os
import time
from urllib.parse import urlparse
import numpy as np

def scrap_page(pages, counter=0):
    
    for page in pages:
        
        print ('starting scrapping ..')
        
        o = urlparse(page)
        
        #create file name to store the scrapped content. file name would contain domain name to differentiate from other scrapped files
        domain = o.netloc
        fn = 'scrapped_text_'+domain+'_'+str(counter)+'.txt'
        scrapped_txt_file = open(fn,'w',encoding='utf-8')
        
        try:
            #Query the website and return the html to the variable 'page'
            page = urllib.request.urlopen(page) #For python 3 use urllib.request.urlopen(wiki)
        except urllib.request.HTTPError:
            print ('got http error skipping this page')
            continue
        #Parse the html in the 'page' variable, and store it in Beautiful Soup format
        scrapped_page_content = BeautifulSoup(page,'html.parser')
        print (BeautifulSoup.prettify(scrapped_page_content))
        
        paras = scrapped_page_content.find_all('p')

        for para in paras:
            item = para.find(text=True)
            if (item is not None):
                scrapped_txt_file.write(item)
                scrapped_txt_file.write('\n')
            
            
    print ('ending scrapping ...')
```

```python
scrap_page(urls_to_be_scrapped)
```
